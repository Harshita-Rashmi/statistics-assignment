1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Descriptive Statistics

Purpose: To summarize and describe the features of a dataset.
What it does:
- Organizes and simplifies raw data
- Uses measures like mean, median, mode, range, standard deviation
- Visualizes data using charts, tables, histograms, box plots
Example: Suppose you collect the test scores of 100 students:
scores = [78, 85, 92, 88, 76, 95, 89, 84, 91, 87]


You might calculate:
- Mean score: sum(scores)/len(scores) → 86.5
- Median score: 87
- Standard deviation: Measures how spread out the scores are
You could also plot a histogram to visualize the distribution.

Inferential Statistics 


Purpose: To make predictions or generalizations about a population based on a sample.
What it does:
- Uses probability theory to estimate population parameters
- Performs hypothesis testing, confidence intervals, regression analysis
- Helps answer questions like: “Is this effect real or just due to chance?”
Example: You survey 100 people about their coffee consumption and find that 60% drink coffee daily. You want to infer whether this holds true for the entire city.
You might:
- Construct a confidence interval (e.g., 60% ± 5%)
- Perform a hypothesis test to see if coffee consumption differs by age group





2.   What is sampling in statistics? Explain the differences between random and stratified sampling. 



Sampling is the process of selecting a subset (sample) from a larger group (population) to analyze and draw conclusions about the whole.
Since it's often impractical to study an entire population (due to time, cost, or accessibility), sampling helps us:
- Estimate population parameters
- Test hypotheses
- Make predictions

- Sampling is the process of selecting a subset from a population to analyze and draw conclusions about the whole.
- It helps reduce cost, time, and effort while still providing meaningful insights.
Random Sampling:
- Every individual in the population has an equal chance of being selected.
- Simple to implement but may miss key subgroups by chance.
- Example: Picking 100 students randomly from a university list.
Stratified Sampling:
- Population is divided into distinct subgroups (strata) based on characteristics like gender, age, or department.
- A random sample is taken from each stratum to ensure balanced representation.
- Example: Selecting 50 males and 50 females from a student population.





3.    Define mean, median, and mode. Explain why these measures of central tendency are important.



1. Mean (Arithmetic Average):
The mean is the sum of all values in a dataset divided by the total number of values.
\text{Mean} = \frac{\sum x_i}{n}
It represents the overall average and is widely used in quantitative analysis.
Example: For scores [10, 20, 30], Mean = (10 + 20 + 30)/3 = 20
2. Median:
The median is the middle value when data is arranged in ascending or descending order.
- If the number of values is odd, it is the central value.
- If even, it is the average of the two middle values.
Example: For [10, 20, 30], Median = 20; for [10, 20, 30, 40], Median = (20+30)/2 = 25
3. Mode:
The mode is the value that appears most frequently in a dataset.
- A dataset can be unimodal, bimodal, or multimodal.
Example: For [10, 20, 20, 30], Mode = 20

 Importance of Measures of Central Tendency
- These measures help summarize large datasets with a single representative value.
- They provide insight into the typical or average behavior of the data.
- Useful in comparing groups, identifying trends, and making informed decisions.
- Each measure has its own strength:
- Mean is ideal for normally distributed data but sensitive to outliers.
- Median is robust against extreme values and useful in skewed distributions.


4.   

Explain skewness and kurtosis. What does a positive skew imply about the data?





Skewness:
- Skewness measures the asymmetry of a data distribution around its mean.
- A perfectly symmetrical distribution (like a normal curve) has zero skewness.
- Positive skew (right-skewed):
- The right tail is longer than the left.
- Most data values are concentrated on the left, with a few high outliers pulling the mean to the right.
- Relationship: Mean > Median > Mode
- Example: Income data, where most earn modest amounts but a few earn very high salaries.
Kurtosis:
- Kurtosis measures the "tailedness" or peakedness of a distribution.
- It indicates how much data is in the tails compared to a normal distribution.
- Types of kurtosis:
- Mesokurtic: Normal distribution (Kurtosis ≈ 3)
- Leptokurtic: Sharp peak, heavy tails (Kurtosis > 3) → More outliers
- Platykurtic: Flat peak, light tails (Kurtosis < 3) → Fewer outliers



- Skewness helps detect asymmetry and potential bias in data.
- Kurtosis helps identify the presence of outliers and assess risk in fields like finance and quality control.
- Together, they provide insight into the shape, spread, and reliability of statistical models.

 What Does a Positive Skew Imply?
- The data has a long tail on the right side.
- Most values are lower, but a few extremely high values stretch the distribution.
- The mean is greater than the median, indicating potential outliers or non-normality.
- Analysts may need to transform the data (e.g., log scale) before applying certain models.

6. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python



list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60] 



import numpy as np

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to NumPy arrays
x = np.array(list_x)
y = np.array(list_y)

# Compute covariance matrix
cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0, 1]

# Compute correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Display results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)

Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.

● Explain how you would use covariance and correlation to explore this relationship. ● Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000]



import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to NumPy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Compute covariance matrix
cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0, 1]

# Compute correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

# Display results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
